# Import all libraries

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
import pandas as pd
 
# Importing the dataset
datas = pd.read_csv('covid_19_world_data - covid_19_world_data.csv')
````

````python
# get all unique countries names
datas["Country/Region"].unique()
````

````python
# get all unique Region  names
datas["WHO Region"].unique()
````

# Parse data 

````python
def calculate_index(index):
  test = []

  for i in range(index):
      test.append([i])

  return np.array(test)

def calculate_value(index):
  test = []

  for i in index:
      test.append([i])

  return np.array(test)
````

To analyze a specific country and a specific region change the country name and region accordingly. Calculations are made according to WHO Region datas. As an example Afghanistan and Europe is used in the code cell below (code cell 9).

In order to get specific datas regarding a specific country/region, use the below given iloc declaration: 



```
iloc[:, 5:6].values --> for confirmed
iloc[:, 6:7].values --> for deaths
iloc[:, 7:8].values --> for recovered
iloc[:, 8:9].values --> for active
```


````python
# change county name to analyse specific country

# makes calculations according to WHO Regions with death data

country = "Afghanistan"
region = "Europe"
x = calculate_value(datas.loc[datas["WHO Region"] == region ].groupby("Date")["Deaths"].mean().values) 
length = len(datas.loc[datas["WHO Region"] == region ].groupby("Date")["Deaths"].sum().values)


y = calculate_index(length)

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33)
````

````python
# change county name to analyse specific country

# makes calculations according to WHO Regions with confirmed data

country = "Afghanistan"
region = "Europe"
x = calculate_value(datas.loc[datas["WHO Region"] == region ].groupby("Date")["Confirmed"].mean().values) 
length = len(datas.loc[datas["WHO Region"] == region ].groupby("Date")["Deaths"].sum().values)


y = calculate_index(length)

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33)
````

````python
# change county name to analyse specific country

# makes calculations according to WHO Regions with active data

country = "Afghanistan"
region = "Europe"
x = calculate_value(datas.loc[datas["WHO Region"] == region ].groupby("Date")["Active"].mean().values) 
length = len(datas.loc[datas["WHO Region"] == region ].groupby("Date")["Deaths"].sum().values)

y = calculate_index(length)

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33)
````

````python
# change county name to analyse specific country

# makes calculations according to WHO Regions with recovered data

country = "Afghanistan"
region = "Europe"
x = calculate_value(datas.loc[datas["WHO Region"] == region ].groupby("Date")["Recovered"].mean().values) 
length = len(datas.loc[datas["WHO Region"] == region ].groupby("Date")["Deaths"].sum().values)


y = calculate_index(length)

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33)
````

To analyze a specific country and a specific region change the country name and region accordingly. As an example Afghanistan and Europe is used in the code cell below (code cell 13).

In order to get specific datas regarding a specific country/region, use the below given iloc declaration: 



```
iloc[:, 5:6].values --> for confirmed
iloc[:, 6:7].values --> for deaths
iloc[:, 7:8].values --> for recovered
iloc[:, 8:9].values --> for active
```



````python
# change county name to analyse specific country

# makes calculations according to countries for death data 

country = "Afghanistan"
region = "Europe"

x = datas.loc[datas["Country/Region"] == country ].iloc[:, 6:7].values # change this line for 4 attribute
length = len(datas.loc[datas["Country/Region"] == country].iloc[:, 6:7].values) # change this line for 4 attribute
y = calculate_index(length)

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33)
````

````python
# change county name to analyse specific country

# makes calculations according to countries for recovered data 

country = "Afghanistan"
region = "Europe"

x = datas.loc[datas["Country/Region"] == country ].iloc[:, 7:8].values # change this line for 4 attribute
length = len(datas.loc[datas["Country/Region"] == country].iloc[:, 7:8].values) # change this line for 4 attribute
y = calculate_index(length)

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33)
````

````python
# change county name to analyse specific country

# makes calculations according to countries for active data 

country = "Afghanistan"
region = "Europe"

x = datas.loc[datas["Country/Region"] == country ].iloc[:, 8:9].values # change this line for 4 attribute
length = len(datas.loc[datas["Country/Region"] == country].iloc[:, 8:9].values) # change this line for 4 attribute
y = calculate_index(length)

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33)
````

# Hyper Parameter Optimization 

Changed the parameters and examined the different results in different parameters below. 

````python
# degree = 4 
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
poly1 = PolynomialFeatures(degree = 4)
X_poly = poly1.fit_transform(X_train)
 
poly1.fit(X_poly, y_train)
lin1 = LinearRegression()
lin1.fit(X_poly, y_train)
````

````python
# degree = 5 
poly2 = PolynomialFeatures(degree = 5)
X_poly2 = poly2.fit_transform(X_train)
 
poly2.fit(X_poly2, y_train)
lin2 = LinearRegression()
lin2.fit(X_poly2, y_train)
````

````python
# degree = 6 
poly3 = PolynomialFeatures(degree = 6)
X_poly3 = poly3.fit_transform(X_train)
 
poly3.fit(X_poly3, y_train)
lin3 = LinearRegression()
lin3.fit(X_poly3, y_train)
````

````python
def calculate_days(days):
  test = []

  for i in range(days):
      test.append([i])

  return np.array(test)
````

# Plotting The Analysis With Graphs 


Here the blue line indicates all the death datas regarding the selected counntry (Afghanistan in our example), orange line indicates the prediction of deaths within 30 days (in Afghanistan)

````python
# plotting for degree =4 
plt.close()
plt.scatter(x, y, color = 'blue')
# np.array([[30]])
# 
plt.scatter(x[:7], lin1.predict(poly1.fit_transform(np.array(calculate_days(7)))), color = 'orange')
plt.title('Polynomial Regression')
plt.xlabel('Time')
plt.ylabel('Death')
print(f"R'2 Score --> {r2_score(y_test, lin1.predict(poly1.fit_transform(X_test)))}")
plt.show()
````

````python


# To show 30 days 

plt.close()
plt.scatter(x, y, color = 'blue')
# for first 30 days use x[:30] and calculate_days(30)
firstDays = 7
plt.scatter(x[:firstDays], lin1.predict(poly1.fit_transform(calculate_days(firstDays))), color = 'orange')
plt.title('Polynomial Regression')
plt.xlabel('Time')
plt.ylabel('Death')
print(f"R'2 Score --> {r2_score(y_test, lin1.predict(poly1.fit_transform(X_test)))}")
plt.show()
````
