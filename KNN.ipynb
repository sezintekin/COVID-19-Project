# Import all libraries

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
 
# Importing the dataset
datas = pd.read_csv('covid_19_world_data - covid_19_world_data.csv')
````

````python
# get all unique countries names
datas["Country/Region"].unique()
````

````python
# get all unique Region  names
datas["WHO Region"].unique()
````

# Parse data 

````python
def calculate_index(index):
  test = []

  for i in range(index):
      test.append([i])

  return np.array(test)

def calculate_value(index):
  test = []

  for i in index:
      test.append([i])

  return np.array(test)
````

To analyze a specific country and a specific region change the country name and region accordingly. Calculations are made according to WHO Region datas. As an example Afghanistan and Europe is used in the code cell below (code cell 14).

In order to get specific datas regarding a specific country/region, use the below given iloc declaration: 



```
iloc[:, 5:6].values --> for confirmed
iloc[:, 6:7].values --> for deaths
iloc[:, 7:8].values --> for recovered
iloc[:, 8:9].values --> for active
```



````python
# change county name to analyse specific country

# makes calculations according to WHO Regions with death data 

country = "Afghanistan"
region = "Europe"
x = calculate_value(datas.loc[datas["WHO Region"] == region ].groupby("Date")["Deaths"].mean().values) 
length = len(datas.loc[datas["WHO Region"] == region ].groupby("Date")["Deaths"].sum().values)

y = calculate_index(length)

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33)
````

````python
# change county name to analyse specific country
# iloc[:, 5:6].values --> for confirmed
# iloc[:, 6:7].values --> for deaths
# iloc[:, 7:8].values --> for recovered
# iloc[:, 8:9].values --> for active

# makes calculations according to countries for active data 

country = "Afghanistan"
region = "Europe"

x = datas.loc[datas["Country/Region"] == country ].iloc[:, 8:9].values # change this line for 4 attribute
length = len(datas.loc[datas["Country/Region"] == country].iloc[:, 8:9].values) # change this line for 4 attribute
y = calculate_index(length)

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33)
````

# Hyper Parameter Optimization 

````python
def calculate_days(days):
  test = []

  for i in range(days):
      test.append([i])

  return np.array(test)
````

Changed the parameters and examined the different results in different parameters below. 

````python
knn1 = KNeighborsClassifier(n_neighbors=4, metric='euclidean')
knn1.fit(X_train, y_train)
````

````python
knn2 = KNeighborsClassifier(n_neighbors=5, metric='euclidean')
knn2.fit(X_train, y_train)
````

````python
knn3 = KNeighborsClassifier(n_neighbors=6, metric='euclidean')
knn3.fit(X_train, y_train)
````

# Plotting the analysis


Here the blue line indicates all the death datas regarding the selected counntry (Afghanistan in our example), red line indicates the prediction of deaths within 30 days (in Afghanistan)

````python
# plotting for n_neighbors=4
plt.close()
plt.scatter(x, y, color = 'blue')
firstDays = 7 
plt.scatter(y_test[:firstDays], knn1.predict(np.array(calculate_days(firstDays))), color = 'red')
plt.title('KNN Algorithm')
plt.xlabel('Time')
plt.ylabel('Death')
print(f"R'2 Score --> {r2_score(y_test, knn1.predict(X_test))}")
plt.show()
````

````python
# plotting for n_neighbors=5
plt.close()
plt.scatter(x, y, color = 'blue')
firstDays = 7
plt.scatter(y_test[:firstDays], knn2.predict(np.array(calculate_days(firstDays))), color = 'red')
plt.title('KNN Algorithm')
plt.xlabel('Time')
plt.ylabel('Death')
print(f"R'2 Score --> {r2_score(y_test, knn2.predict(X_test))}")
plt.show()
````

````python
# plotting for  n_neighbors=6
plt.close()
firstDays = 7
plt.scatter(x, y, color = 'blue')
plt.scatter(y_test[:firstDays], knn3.predict(np.array(calculate_days(firstDays))), color = 'red')
plt.title('KNN Algorithm')
plt.xlabel('Time')
plt.ylabel('Death')
print(f"R'2 Score --> {r2_score(y_test, knn3.predict(X_test))}")
plt.show()
````
